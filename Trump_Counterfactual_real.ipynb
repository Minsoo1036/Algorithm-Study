```python
import datetime
from dateutil.relativedelta import relativedelta

def to_date(date_in, date_form = "%Y-%m-%d"):
    """
    datetime.datetime()이나 str을 datetime.date()화
    """
    if type(date_in) == str and len(date_in.split("-")) == 3:
        return(datetime.datetime.strptime(date_in, date_form).date())
    else:
        try:
            return(date_in.date())
        except:
            return(date_in)
```

```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(42)  # Set the seed number to ensure reproducibility

# Step 1: Prepare your quarterly data
#data = {'ds': ['2020-01-01', '2020-04-01', '2020-07-01', '2020-10-01',
#               '2021-01-01', '2021-04-01', '2021-07-01', '2021-10-01'],
#        'y': [100, 150, 200, 250, 120, 170, 220, 270]}
#df = pd.DataFrame(data)
#df['ds'] = pd.to_datetime(df['ds'])  # Ensure 'ds' is datetime

# Step 2: Initialize and fit the model with quarterly seasonality
m = Prophet(growth='linear',
            changepoints = None,
            changepoint_range = 0.90,
            changepoint_prior_scale = 0.9,
            seasonality_prior_scale = 10,
            daily_seasonality = False,
            weekly_seasonality=False,
            yearly_seasonality=False,
            seasonality_mode = 'additive',
            interval_width = 0.7,
            mcmc_samples = 100)
m.add_seasonality(name='quarterly', period=4, fourier_order=5)
m.fit(df)

# Step 3: Create future dataframe for next 8 quarters
# future = m.make_future_dataframe(periods=8,freq='Q')

future = m.make_future_dataframe(periods=8,freq='Q', include_history=False)
future['ds'] = future['ds'].apply(lambda x: x + relativedelta(days=1))
future = pd.concat([df['ds'], future]).apply(lambda x: pd.to_datetime(x))

# Step 4: Forecast future points
forecast = m.predict(future)

# Step 5: Plot forecasted values
fig1 = m.plot(forecast)
plt.show()

# Step 6: Plot the components (including quarterly seasonality)
fig2 = m.plot_components(forecast)
plt.show()
```
